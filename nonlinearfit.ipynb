```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

# Define the Hicks equation as a Python function
def hicks_equation(Re_P, alpha, gamma):
    return alpha * Re_P**gamma

initial_alpha_guess = 6.8
initial_gamma_guess = -0.2

# Updated experimental data for Reynolds number and friction factor
Re_P_combined = np.array([2521.354081, 1525.502575, 711.0401837, 1716.168955, 1049.917663, 2203.028817, 2357.254414, 1875.118953, 878.1131491, 2035.362621, 1368.671767, 1205.84433])
friction_factor_combined = np.array([1.546637557, 1.719865013, 2.105205575, 1.667615016, 1.836235167, 1.58279576, 1.574272811, 1.648694229, 1.950611352, 1.618491171, 1.756665271, 1.813905847])

# Perform the regression analysis for the combined dataset
params_combined, covariance_combined = curve_fit(hicks_equation, Re_P_combined, friction_factor_combined, p0=(initial_alpha_guess, initial_gamma_guess))

# Extract the estimated parameters for the combined dataset
alpha_estimate_combined, gamma_estimate_combined = params_combined

# Plot the experimental data and regression model for the combined dataset
plt.scatter(Re_P_combined, friction_factor_combined, label='Experimental Data (Combined)')
plt.plot(Re_P_combined, hicks_equation(Re_P_combined, alpha_estimate_combined, gamma_estimate_combined), label='Regression Model (Combined)', color='green')

# Print the estimated values of alpha and gamma for the combined dataset
print("Estimated alpha (Combined):", alpha_estimate_combined)
print("Estimated gamma (Combined):", gamma_estimate_combined)

# Show the plot
plt.legend()
plt.xlabel('Reynolds Number (Re_P)')
plt.ylabel('Friction Factor (f)')
plt.show()
```
